**คำอธิบาย**
วิธีการของนิวตันเป็นวิธีการที่ใช้ประมาณค่าของนิพจน์ใด ๆ โดยใช้หลักการการลากเส้นสัมผัสกับเส้นกราฟ แต่ในโจทย์ข้อนี้ นิสิตไม่จำเป็นต้องรู้ถึงที่มาที่ไปของวิธีการนี้

นิสิตสามารถหาค่าประมาณของคำตอบของสมการใด ๆ ได้ด้วยการหาค่าของ $x_n$ ที่เข้าใกล้ค่าที่ต้องการได้ ตามสมการด้านล่างนี้

$x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)}$

เช่น นิสิตต้องการหาค่าประมาณของรากที่สองของ $a$ นิสิตจะหาค่าของพจน์ต่าง ๆ ได้ดังนี้

$f(x)=x^2-a=0\\
f'(x)=2x$

หากกำหนดให้ $x_0$ เป็นค่าเริ่มต้นใด ๆ เช่น 20 และต้องการหาค่ารากที่สองของ $a=1000$ นิสิตจะสามารถหา $x_1$ ได้ด้วยสมการ

$x_1=x_0-\frac{f(x_0)}{f'(x_0)}=20-\frac{20^2-1000}{2\times 20}=35.0$ <br>
$x_2=x_1-\frac{f(x_1)}{f'(x_1)}=35.0-\frac{35.0^2-1000}{2\times 35}=31.785714285714285$ <br>
$x_3=x_2-\frac{f(x_2)}{f'(x_2)}=\dots $

ค่า $x_n$ ในพจน์ถัด ๆ ไป จะมีค่าเข้าใกล้รากที่สองของ 1000 เข้าไปเรื่อย ๆ ซึ่งเราสามารถหาค่าความใกล้ได้โดย $|x_n^2 - 1000|$

ตัวอย่างเช่น

ค่าความใกล้ของ $x_1$ จะเท่ากับ $|35.0^2 - 1000|=225.0$ <br>
ค่าความใกล้ของ $x_2$ จะเท่ากับ $|31.785714285714285^2 - 1000|=10.331632653061206$


ให้นิสิตเขียนโปรแกรมเพื่อหาค่ารากที่ $k$ ของโจทย์ที่กำหนด โดย **ไม่ให้ใช้** Loop ให้นิสิตหาว่าต้องคำนวณถึง $n$ ใน $x_n$ ที่เท่าใด จึงจะได้คำตอบที่ค่าความใกล้ต่ำกว่า $10^{-6}$

**การตรวจให้คะแนน**

## โจทย์ของนิสิต

ให้นิสิตป้อนรหัสนิสิต นิสิตจะเห็นโจทย์ของนิสิตปรากฏขึ้นด้านล่างของเซลล์นี้ ให้เขียนโปรแกรมในเซลล์ถัดไป ***เท่านั้น*** โดยเอาต์พุตที่ต้องการคือการพิมพ์ค่า $n$ เพียงอย่างเดียว

```python
# @title ป้อนรหัสนิสิตเพื่อสร้างโจทย์ของตนเอง
student_id = "6632007521" # @param {type:"string"}

import random
random.seed(31+int(student_id))
k=int(random.random()*8)+2
a=int(random.random()*1000)+100
x0=int(random.random()*(a-1))+1

print(f'ให้นิสิตเขียนโปรแกรมเพื่อหาค่ารากที่ {k} ของ {a} โดยไม่ใช้ Loop ให้พิมพ์ค่า n น้อยที่สุดที่ทำให้ได้คำตอบที่ค่าความใกล้ต่ำกว่า 10E−6 โดยกำหนดให้ค่าเริ่มต้น (x0) เป็น {x0}')
```
